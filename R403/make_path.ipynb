```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_shift(current_p, next_p):
    shift = next_p - current_p
#     if np.abs(shift[0]) == np.abs(shift[1]):
#         shift = shift / np.abs(shift + 1e-10)
#         shift = np.round(shift).astype(int)
#     else:
#         distance = np.sqrt(np.sum(np.power(shift, 2)))
#         shift = shift / distance
    distance = np.round(np.sqrt(np.sum(np.power(shift, 2))))
    shift = shift / distance

    return shift


def make_path(points):
    if len(points) <= 1:
        raise
        
    path = []
    points = np.array(points)
    
    for i in range(len(points) - 1):
        current_p = points[i]
        next_p = points[i + 1]
        
        shift = find_shift(current_p, next_p)
        
        start_p = current_p
        
#         print(f'shift: {shift}')
        while (np.round(start_p, decimals=2) != next_p).any():
#             print(start_p)
            path.append(start_p)
            start_p = start_p + shift
    path.append(next_p)
    
    return np.array(path)
```

```python
def plot_path(paths, title):
    plt.figure()
    plt.scatter(grids[:, 0], grids[:, 1], c='lightgray')

    for i, path in enumerate(paths):
        plt.plot(path[:, 1], path[:, 0], label=f'p{i}', alpha=0.75)
        plt.scatter(path[0, 1], path[0, 0], c='tab:green', alpha=0.75)
        plt.scatter(path[-1, 1], path[-1, 0], c='tab:red', alpha=0.75)

    plt.scatter(path[0, 1], path[0, 0], c='tab:green', alpha=0.75, label='start')
    plt.scatter(path[-1, 1], path[-1, 0], c='tab:red', alpha=0.75, label='end')
    plt.title(title)
    plt.axis('equal')
    plt.legend()
    plt.show()
```
